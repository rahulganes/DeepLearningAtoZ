```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset = pd.read_csv("Credit_Card_Applications.csv")
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:, -1].values
```

```python
sc = MinMaxScaler(feature_range=(0,1))
X = sc.fit_transform(X)
```

<font color = 'green' size="+2"><b>MiniSom</b></font>
         <font size = "+1"><br>&emsp;&emsp;&emsp;x and y are dimensions of grid in SOM, their values of arbitary.The i/p length depends to number of features in dataframe x.sigma is the radius of neighbourhood.learning rate decides the weight updated.><\font>

```python
from minisom import MiniSom
som = MiniSom(x=10,y=10,input_len = 15, sigma = 1.0, learning_rate = 0.5)
som.random_weights_init(X)
som.train_random(data = X, num_iteration =100)
```

```python
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o','s']
colors = ['r','g']
for i,x in  enumerate(X):
    w = som.winner(x)
    plot(w[0]+0.5,w[1]+0.5,markers[Y[i]],markeredgecolor=colors[Y[i]],markerfacecolor='None',markersize=10,markeredgewidth=2)
show()
```
